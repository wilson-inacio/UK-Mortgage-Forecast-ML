# End-to-End Machine Learning Pipeline for Forecasting UK Mortgage Rates
## Overview
UK mortgage rates directly impact affordability, refinancing decisions, and lender risk management. The objective of this project is to **build a machine learning pipeline that forecasts short-term mortgage rate movements** to support decisions such as refinancing timing, product pricing, and risk assessment.

The project demonstrates a complete end-to-end workflow from data acquisition to model evaluation and can be converted into a production-ready pipeline, focusing on short-term montly forecasts.

## 1. Data Loading & Preprocessing
In this section, we'll import the relevant libraries and load CSVs, clean column names, convert dates and align monthly data, finalising with merging all the dataframes into one so we can proceed with our analysis.

```python
# Loading libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split, TimeSeriesSplit, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import joblib
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Loading the dataframes
fixed_2y = pd.read_csv('resources/data/2y_fixed.csv')
fixed_5y = pd.read_csv('resources/data/5y_fixed.csv')
base_rate = pd.read_csv('resources/data/BoE_base_rate.csv')
cpih = pd.read_csv('resources/data/Inflation(CPIH).csv')

# Confirming load
display(fixed_2y.head())
display(fixed_5y.head())
display(base_rate.head())
display(cpih.head(10))
```

From this initial assessment, it's noticeable that some care has to be had with the column names, which for the two dataframes are too long. The proper data on the dataframe `cpih` only starts from row 7, with generic information on the first 6 rows, which can be safely removed.

```python
# Renaming the necessary columns
fixed_2y = fixed_2y.rename(columns = {fixed_2y.columns[1]: '95_LTV_2y', 
                                    fixed_2y.columns[2]: '90_LTV_2y',
                                    fixed_2y.columns[3]: '75_LTV_2y',
                                    fixed_2y.columns[4]: '60_LTV_2y'})
fixed_5y = fixed_5y.rename(columns = {fixed_5y.columns[1]: '95_LTV_5y',
                                    fixed_5y.columns[2]: '75_LTV_5y',
                                    fixed_5y.columns[3]: '60_LTV_5y',
                                    fixed_5y.columns[4]: '90_LTV_5y'})
cpih = cpih.rename(columns = {cpih.columns[0]: 'Date', cpih.columns[1]: 'CPIH'})

# Removing the first 7 rows of cpih
cpih = cpih.drop(index = cpih.index[:7]).reset_index(drop = True)

display(fixed_2y.head(3))
display(fixed_5y.head(3))
display(base_rate.head(3))
display(cpih.head(3))
```

Now that our dataframes are starting to gain shape, we'll proceed by running `info()` on the dataframes, as that lets us see if there are missing entries and also if the column data types are aligned with what would be expected.

```python
# Checking information on the dataframes
print('\nFixed 2 years:')
print(fixed_2y.info())
print('\nFixed 5 years:')
print(fixed_5y.info())
print('\nBase Rate:')
print(base_rate.info())
print('\nPCIH:')
print(cpih.info())
```

It seems that all of the `Date` columns are appearing as object `type`, so we'll need to convert them to datetime objects, and also looking at the fixed-rate dataframes, some columns that were expected to be `float` are currently `object`, this means that there might be NaN or non-numeric values, so we'll need to verify. 

```python
# Converting all columns to numeric and counting how many values are either NaN or non-numeric
rate_cols_2y = fixed_2y.columns.drop('Date')
coerced_2y = fixed_2y[rate_cols_2y].apply(pd.to_numeric, errors="coerce")
missing_mask_2y = coerced_2y.isna() & fixed_2y.notna()
print('2 Year fixed-rate:\n', missing_mask_2y.sum())

rate_cols_5y = fixed_5y.columns.drop('Date')
coerced_5y = fixed_5y[rate_cols_5y].apply(pd.to_numeric, errors="coerce")
missing_mask_5y = coerced_5y.isna() & fixed_5y.notna()
print('5 Year fixed-rate:\n', missing_mask_5y.sum())

# Joining it all back again
fixed_2y[rate_cols_2y] = coerced_2y
fixed_5y[rate_cols_5y] = coerced_5y
```

Numeric coercion was applied to all mortgage rate series. Values failing conversion while remaining present in the raw data were identified as placeholder entries denoting unavailable observations and were subsequently treated as missing during preprocessing.<br>
Among available LTV bands, both 60% and 75% LTV series were fully observed across 2-year and 5-year products, but **the 75% LTV series was selected as the primary modelling target** due to its widespread use as a market benchmark and greater sensitivity to macroeconomic conditions.

The next step is then to clean the `Date` columns and format them the same way across all dataframes, this allows proper time-aware processing, indexing, and alignment across datasets. The CPIH series required a custom format due to "YYY MMM" strings.

```python
# Converting all `Dates` to datetime
fixed_2y['Date'] = fixed_2y['Date'].apply(pd.to_datetime, dayfirst = True)
fixed_5y['Date'] = fixed_5y['Date'].apply(pd.to_datetime, dayfirst = True)
base_rate['Date'] = base_rate['Date'].apply(pd.to_datetime, dayfirst = True)
cpih['Date'] = cpih['Date'].apply(pd.to_datetime, format='%Y %b')

# Confirming changes applied as predicted
display(fixed_2y.head(3))
print(fixed_2y.dtypes)
```

We can now group `base_rate` by month and then remove all the days from the `Date` column, keeping the formatting the same across all dataframes.

```python
# Grouping by Month
base_rate_monthly = base_rate.groupby(base_rate['Date'].dt.to_period('M').rename('Month')).mean().reset_index()

# Convert Period back to datetime at start of month
base_rate_monthly['Date'] = base_rate_monthly['Month'].dt.to_timestamp()
base_rate_monthly.drop(columns='Month', inplace=True)

# Dropping day from all dataframes
fixed_2y['Date'] = fixed_2y['Date'].dt.to_period('M').dt.to_timestamp()
fixed_5y['Date'] = fixed_5y['Date'].dt.to_period('M').dt.to_timestamp()
base_rate_monthly['Date'] = base_rate_monthly['Date'].dt.to_period('M').dt.to_timestamp()
cpih['Date'] = cpih['Date'].dt.to_period('M').dt.to_timestamp()


display(base_rate_monthly.head())
display(fixed_2y.head())
display(fixed_5y.head())
cpih.head()
```

The preprocessing for our datasets is almost complete, the last step is to join all the dataframes into a single one with all the information necessary.<br>
The datasets will be merged on month-start dates, retaining only the target series (75% LTV mortgage rate) and relevant macroeconomic features (`Bank Rate` as it directly influences lender pricing, and `CPIH` which reflects inflationary pressure). This produces a single, time-aligned table for modelling.<br>
**Two dataframes were created** so we can proceed with the **analysis of both a 2-year and a 5-year fixed rate** for mortgages with 75% LTV without risking information being leaked between the rates or compromising the predictions. An inner join was used to ensure all series share a common observation window, preventing implicit forward-filling or extrapolation across datasets.

```python
# Joining all the dataframes / Inner was selected to make sure the time period is the same across all columns
# 2 year fixed
df_2y = fixed_2y[['Date', '75_LTV_2y']].merge(
    base_rate_monthly[['Date', 'Bank Rate']],
    on='Date',
    how='inner'
).merge(
    cpih[['Date', 'CPIH']],
    on='Date',
    how='inner'
)
# 5 year fixed
df_5y = fixed_5y[['Date', '75_LTV_5y']].merge(
    base_rate_monthly[['Date', 'Bank Rate']],
    on='Date',
    how='inner'
).merge(
    cpih[['Date', 'CPIH']],
    on='Date',
    how='inner'
)

# Sorting the Dated from older to newer
df_2y = df_2y.sort_values('Date')
df_5y = df_5y.sort_values('Date')

# Setting the `Date` as the index so resampling and plotting are easier 
df_2y.set_index('Date', inplace=True)
df_5y.set_index('Date', inplace=True)

# Confirming
print(df_2y.head())
print(df_5y.info())
```

## 2. Feature Engineering 
We'll now proceed with creating lagged mortgage rates so we can capture the persistence of the housing market in relation to current market trends and a short-term rolling mean to capture short-term trends, and finally create our feature and target variables. <br>
On this project, two target variables will be used, one for a fixed 2-year fixed rate with 75% LTV and another for a fixed 5-year fixed rate with 75% LTV.

```python
# Lagged Mortgage rates
# 2 year
df_2y['mortgage_lag_1'] = df_2y['75_LTV_2y'].shift(1)
df_2y['mortgage_lag_3'] = df_2y['75_LTV_2y'].shift(3)

# 5 year
df_5y['mortgage_lag_1'] = df_5y['75_LTV_5y'].shift(1)
df_5y['mortgage_lag_3'] = df_5y['75_LTV_5y'].shift(3)

# Rolling mean
df_2y['mortgage_roll_3'] = df_2y['75_LTV_2y'].rolling(window=3).mean()
df_5y['mortgage_roll_3'] = df_5y['75_LTV_5y'].rolling(window=3).mean()

# Drop rows with NaNs created by the lagged rates
df_2y_fe = df_2y.dropna().copy()
df_5y_fe = df_5y.dropna().copy()

# Creating feature and target dataframes
# 2 year
X_2y = df_2y_fe[['Bank Rate', 'CPIH', 'mortgage_lag_1', 'mortgage_lag_3', 'mortgage_roll_3']]
y_2y = df_2y_fe['75_LTV_2y']

# 5 year
X_5y = df_5y_fe[['Bank Rate', 'CPIH', 'mortgage_lag_1', 'mortgage_lag_3', 'mortgage_roll_3']]
y_5y = df_5y_fe['75_LTV_5y']
```

## 3. Train/Test Split
Train-test splits were performed after feature construction to preserve the temporal structure of lagged variables. To avoid future data compromising the trends and random shuffling breaking causality, a time-aware split was used, preventing look-ahead bias.

```python
# Doing the train-test splits
X_train_2y, X_test_2y, y_train_2y, y_test_2y = train_test_split(X_2y, y_2y, test_size = 0.2, shuffle = False)
X_train_5y, X_test_5y, y_train_5y, y_test_5y = train_test_split(X_5y, y_5y, test_size = 0.2, shuffle = False)
```

## 4. Baseline Model: Linear Regression
For baseline models, two will be compared, the first one, a naive model, is used, which is already in place in the feature `mortgage_lag_1`. The second model will be a simple linear regression, allowing us to see how the dataframe features impact the prediction. As evaluation metric, the **Mean Absolute Error (MAE)** will be used, as it **directly reflects the average absolute error in mortgage rate percentage points**, which is meaningful for pricing and refinancing decisions. Root Mean Squared Error (RMSE) is used as a complementary metric to further help distinguish and evaluate the models.

```python
# Setting the Naive dataframes
y_pred_naive_2y = X_test_2y['mortgage_lag_1']
y_pred_naive_5y = X_test_5y['mortgage_lag_1']

# Training the models
lr_model_2y = LinearRegression()
lr_model_2y.fit(X_train_2y, y_train_2y)

lr_model_5y = LinearRegression()
lr_model_5y.fit(X_train_5y, y_train_5y)

# Making predictions
y_pred_lr_2y = lr_model_2y.predict(X_test_2y)
y_pred_lr_5y = lr_model_5y.predict(X_test_5y)

# Evaluating the models
print('Naive Baseline Evaluation for 2 year fixed rate:')
print('MAE:', mean_absolute_error(y_test_2y, y_pred_naive_2y))
print('RMSE:', np.sqrt(mean_squared_error(y_test_2y, y_pred_naive_2y)))
print('\n')
print('Naive Baseline Evaluation for 5 year fixed rate:')
print('MAE:', mean_absolute_error(y_test_5y, y_pred_naive_5y))
print('RMSE:', np.sqrt(mean_squared_error(y_test_5y, y_pred_naive_5y)))
print('\n')
print('Evaluation for 2 year fixed rate model:')
print('MAE:', mean_absolute_error(y_test_2y, y_pred_lr_2y))
print('RMSE:', np.sqrt(mean_squared_error(y_test_2y, y_pred_lr_2y)))
print('\n')
print('Evaluation for 5 year fixed rate model:')
print('MAE:', mean_absolute_error(y_test_5y, y_pred_lr_5y))
print('RMSE:', np.sqrt(mean_squared_error(y_test_5y, y_pred_lr_5y)))

# Visual representation
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,8), sharey = True)
fig.suptitle('Linear Regression/Naive Model')
ax[0,0].plot(y_test_2y.index, y_test_2y, label='Actual')
ax[0,0].plot(y_test_2y.index, y_pred_naive_2y, label='Predicted', linestyle = '--')
ax[0,0].legend()
ax[0,0].set_title('Naive Model: 2Y Mortgage Rate: Actual vs Predicted')
ax[0,0].tick_params(axis='x', rotation=45)

ax[0,1].plot(y_test_5y.index, y_test_5y, label='Actual')
ax[0,1].plot(y_test_5y.index, y_pred_naive_5y, label='Predicted', linestyle = '--')
ax[0,1].legend()
ax[0,1].set_title('Naive Model: 5Y Mortgage Rate: Actual vs Predicted')
ax[0,1].tick_params(axis='x', rotation=45)

ax[1,0].plot(y_test_2y.index, y_test_2y, label='Actual')
ax[1,0].plot(y_test_2y.index, y_pred_lr_2y, label='Predicted', linestyle = '--')
ax[1,0].legend()
ax[1,0].set_title('Linear Regression: 2Y Mortgage Rate: Actual vs Predicted')
ax[1,0].tick_params(axis='x', rotation=45)

ax[1,1].plot(y_test_5y.index, y_test_5y, label='Actual')
ax[1,1].plot(y_test_5y.index, y_pred_lr_5y, label='Predicted', linestyle = '--')
ax[1,1].legend()
ax[1,1].set_title('Linear Regression: 5Y Mortgage Rate: Actual vs Predicted')
ax[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

An **MAE for both variables close to 0.10** means that the model predicts the fixed mortgage rates with an error of up to 0.11 and 0.10 percentage points. The values of **RMSE of close to 0.03** indicate that the predictions are very close to the actual, observed data points. **Both metrics confirm that the predictions are fairly accurate.**

## 5. Machine Learning Model: Random Forest
For the machine learning model of this project, a Random Forest model was selected, as it can capture non-linear relationships, it's robust to small datasets, and it is suitable for tabular macroeconomic time-series features without extensive preprocessing.<br>
Hence, as on this project the **Linear Regression model is being used as a baseline, Random Forest will be used as a non-linear benchmark.**

```python
# Implementing the RandomForestRegressor
rfr_2y = RandomForestRegressor(random_state = 30,
                            n_estimators = 300,
                            max_depth = 5,
                            min_samples_leaf = 5)
rfr_5y = RandomForestRegressor(random_state = 30,
                            n_estimators = 300,
                            max_depth = 5,
                            min_samples_leaf = 5)

# Fitting and testing
rfr_2y.fit(X_train_2y, y_train_2y)
rfr_5y.fit(X_train_5y, y_train_5y)

y_pred_rf_2y = rfr_2y.predict(X_test_2y)
y_pred_rf_5y = rfr_5y.predict(X_test_5y)

# Evaluating the models
print('Evaluation for 2 year fixed rate model:')
print('MAE:', mean_absolute_error(y_test_2y, y_pred_rf_2y))
print('RMSE:', np.sqrt(mean_squared_error(y_test_2y, y_pred_rf_2y)))
print('\n')
print('Evaluation for 5 year fixed rate model:')
print('MAE:', mean_absolute_error(y_test_5y, y_pred_rf_5y))
print('RMSE:', np.sqrt(mean_squared_error(y_test_5y, y_pred_rf_5y)))

# Visual representation
fig, (ax2, ax5) = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4), sharey = True)
fig.suptitle('Random Forest Model')
ax2.plot(y_test_2y.index, y_test_2y, label='Actual')
ax2.plot(y_test_2y.index, y_pred_rf_2y, label='Predicted', linestyle = '--')
ax2.legend()
ax2.set_title('2Y Mortgage Rate: Actual vs Predicted')
ax2.tick_params(axis='x', rotation=45)

ax5.plot(y_test_5y.index, y_test_5y, label='Actual')
ax5.plot(y_test_5y.index, y_pred_rf_5y, label='Predicted', linestyle = '--')
ax5.legend()
ax5.set_title('5Y Mortgage Rate: Actual vs Predicted')
ax5.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

Random Forest regressors were trained separately for 2-year and 5-year fixed-rate targets, using identical hyperparameters to ensure comparability across horizons, and were trained with constrained depth to reduce overfitting while capturing non-linear relationships between macroeconomic indicators and mortgage rates. Model randomness was controlled for reproducibility.<br>
The **values achieved** when computing the evaluation **were worse for both the MAE and the RMSE**, meaning that despite its flexibility, the **Random Forest underperformed** relative to the linear baseline. This can also be checked visually through the plot, as the predicted values in the Linear Regression Model more closely follow the actual values, when compared to the Random Forest Model. This suggests that the mortgage rates over the observed period are well approximated by linear relationships with macroeconomic variables and that additional model complexity introduces variance without improving predictive accuracy.

For the Random Forest model, some of its hyperparamenters will now be fine-tuned, as to verify if even with optimal values for this model, it still underperforms when compared to the Linear Regression Model.

```python
# Defining parameters to fine-tune
param_grid = {'n_estimators': [200, 500],
              'max_depth': [3, 5, 10, None],
              'min_samples_leaf': [1, 3, 5],
              'max_features': ['sqrt', 0.7]
             }

# Setting split to be used in GridSearch
'''
As SearchGridCV searches randomly, data may leak from future to past and influence predictions.
Due to this, need a way to do the gridsearch in a way that follows cronological order. >> use time series split
(Similar to the train test split without shuffling, used before, but this makes it so instead of just a split at
0.8/0.2, it makes different splits for each iteraction
'''
tscv = TimeSeriesSplit(n_splits=5)

# GridSearch for 2y model
rf_2y = RandomForestRegressor(random_state=42)

grid_2y = GridSearchCV(estimator=rfr_2y,
                       param_grid=param_grid,
                       cv=tscv,
                       scoring='neg_mean_absolute_error',
                       n_jobs=-1
                      )

grid_2y.fit(X_train_2y, y_train_2y)

best_rf_2y = grid_2y.best_estimator_

# Making prediction
y_pred_rf_tuned_2y = best_rf_2y.predict(X_test_2y)

# GridSearch for 5y model
rf_5y = RandomForestRegressor(random_state=42)

grid_5y = GridSearchCV(estimator=rfr_5y,
                       param_grid=param_grid,
                       cv=tscv,
                       scoring='neg_mean_absolute_error',
                       n_jobs=-1
                      )

grid_5y.fit(X_train_5y, y_train_5y)

best_rf_5y = grid_5y.best_estimator_

# Making prediction
y_pred_rf_tuned_5y = best_rf_5y.predict(X_test_5y)

# Evaluating the models
mae_2y = mean_absolute_error(y_test_2y, y_pred_rf_tuned_2y)
rmse_2y = np.sqrt(mean_squared_error(y_test_2y, y_pred_rf_tuned_2y))
mae_5y = mean_absolute_error(y_test_5y, y_pred_rf_tuned_5y)
rmse_5y = np.sqrt(mean_squared_error(y_test_5y, y_pred_rf_tuned_5y))

print(f"Tuned RF  2Y MAE: {mae_2y}")
print(f"Tuned RF  2Y RMSE: {rmse_2y}")
print(f"Tuned RF  5Y MAE: {mae_5y}")
print(f"Tuned RF  5Y RMSE: {rmse_5y}")

# Plotting results
fig, (ax2, ax5) = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4), sharey = True)

ax2.plot(y_test_2y.index, y_test_2y, label='Actual')
ax2.plot(y_test_2y.index, y_pred_rf_tuned_2y, label='RF (tuned)', linestyle='--')
ax2.legend()
ax2.set_title('2Y Mortgage Rate: Actual vs Predicted')
ax2.tick_params(axis='x', rotation=45)


ax5.plot(y_test_5y.index, y_test_5y, label='Actual')
ax5.plot(y_test_5y.index, y_pred_rf_tuned_5y, label='RF (tuned)', linestyle='--')
ax5.legend()
ax5.set_title('5Y Mortgage Rate: Actual vs Predicted')
ax5.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

While a single train-test split preserves temporal order, **TimeSeriesSplit provides a more robust evaluation by validating model performance across multiple historical forecasting windows**, so it was used during hyperparameter tuning to preserve the temporal ordering of the data and prevent look-ahead bias, ensuring that model evaluation reflects a realistic forecasting setting.<br>
**Hyperparameter tuning marginally improved Random Forest performance** but did not close the gap with the linear regression benchmark. This suggests that **the relationship between mortgage rates and macroeconomic drivers is largely linear** over the observed period and that tree-based models introduce unnecessary complexity. We can deduce mortgage rates are primarily driven by stable, approximately linear macroeconomic relationships rather than complex nonlinear interactions.

## 7. Machine Learning Pipeline
Now that feature engineering, train/test split and model evaluation are done, we can proceed to create a pipeline that modularises everything.

```python
# Defining preprocessing and model
pipeline_2y = Pipeline([('Scaler', StandardScaler()),
                       ('lr_model', LinearRegression())])

pipeline_5y = Pipeline([('Scaler', StandardScaler()),
                       ('lr_model', LinearRegression())])

# Fitting the pipeline
pipeline_2y.fit(X_train_2y, y_train_2y)
pipeline_5y.fit(X_train_5y, y_train_5y)

# Making predictions
y_pred_pipeline_2y = pipeline_2y.predict(X_test_2y)
y_pred_pipeline_5y = pipeline_5y.predict(X_test_5y)

# Evaluating the pipeline models
mae_2y = mean_absolute_error(y_test_2y, y_pred_pipeline_2y)
rmse_2y = np.sqrt(mean_squared_error(y_test_2y, y_pred_pipeline_2y))
print("Pipeline MAE for 2 Year Fixed Rate:", mae_2y)
print("Pipeline RMSE for 2 Year Fixed Rate:", rmse_2y)

mae_5y = mean_absolute_error(y_test_5y, y_pred_pipeline_5y)
rmse_5y = np.sqrt(mean_squared_error(y_test_5y, y_pred_pipeline_5y))
print("Pipeline MAE for 5 Year Fixed Rate:", mae_5y)
print("Pipeline RMSE for 5 Year Fixed Rate:", rmse_5y)

# Saving Pipelines for possible future use
joblib.dump(pipeline_2y, 'resources/models/linear_regression_2y_pipeline.pkl')
joblib.dump(pipeline_5y, 'resources/models/linear_regression_5y_pipeline.pkl')

''' 
Usage:
pipeline_loaded = joblib.load('linear_regression_2y_pipeline.pkl')
next_month_pred = pipeline_loaded.predict(next_month_features)
'''
```

## 8. Model Saving & Inference Function
On this step, the trained models will be saved onto a file, and a reusable prediction function that loads the desired model and a newer dataframe and returns the predicted values.

```python
# Save the trained models
joblib.dump(lr_model_2y, "resources/models/linear_regression_2y.pkl")
joblib.dump(lr_model_5y, "resources/models/linear_regression_5y.pkl")

# Create a reusable prediction function
def predict_mortgage(pipeline_path, new_features):
    """
    Load a saved pipeline and predict mortgage rates for new features.

    Args:
        pipeline_path (str): Path to the saved pipeline.
        new_features (pd.DataFrame): DataFrame with the same columns as training features.

    Returns:
        np.ndarray: Predicted mortgage rates.
    """
    pipeline = joblib.load(pipeline_path)
    predictions = pipeline.predict(new_features)
    return predictions
```

The code above allows us to call the function with an up-to-date dataframe that should contain the same columns as used in our dataframes throughout the project (`Bank Rate`, `CPIH`, `mortgage_lag_1`, `mortgage_lag_3` and `mortgage_roll_3`), and the value to predict (2 Year Fixed Rate Interest or 5 Year Fixed Rate Interest) and returns the predicted values based on the new dataframe.<br>
Using this function, we can calculate the rates for the next 2 months:

```python
# 2 year fixed rate forecast
pipeline_2y = joblib.load('resources/models/linear_regression_2y_pipeline.pkl')

last_rows = df_2y_fe.iloc[-3:].copy()
future_preds_2y = []

for _ in range(2):  # next 2 months
    
    next_month_features = pd.DataFrame({
        'Bank Rate': [3.75],
        'CPIH': [3.4],
        'mortgage_lag_1': [last_rows['75_LTV_2y'].iloc[-1]],
        'mortgage_lag_3': [last_rows['75_LTV_2y'].iloc[0]],
        'mortgage_roll_3': [last_rows['75_LTV_2y'].mean()]
    })

    # Predict
    pred = pipeline_2y.predict(next_month_features)[0]
    
    # Store prediction
    future_preds_2y.append(pred)
    
    # Update rolling window
    last_rows = pd.concat(
        [last_rows, pd.DataFrame({'75_LTV_2y': [pred]})],
        ignore_index=True
    ).iloc[-3:]

print("Predicted next 2 months 2Y fixed-rate:",
      [float(round(x, 2)) for x in future_preds_2y])


# 5 year fixed rate forecast
pipeline_5y = joblib.load('resources/models/linear_regression_5y_pipeline.pkl')

last_rows = df_5y_fe.iloc[-3:].copy()
future_preds_5y = []

for _ in range(2):  # next 2 months
    
    next_month_features = pd.DataFrame({
        'Bank Rate': [3.75],
        'CPIH': [3.4],
        'mortgage_lag_1': [last_rows['75_LTV_5y'].iloc[-1]],
        'mortgage_lag_3': [last_rows['75_LTV_5y'].iloc[0]],
        'mortgage_roll_3': [last_rows['75_LTV_5y'].mean()]
    })

    # Predict
    pred = pipeline_5y.predict(next_month_features)[0]
    
    # Store prediction
    future_preds_5y.append(pred)
    
    # Update rolling window
    last_rows = pd.concat(
        [last_rows, pd.DataFrame({'75_LTV_5y': [pred]})],
        ignore_index=True
    ).iloc[-3:]

print("Predicted next 2 months 5Y fixed-rate:",
      [float(round(x, 2)) for x in future_preds_5y])
```

#### Conclusion
This project successfully developed and evaluated predictive models for UK 2-year and 5-year fixed-rate mortgage products using historical economic indicators. Through systematic experimentation and model comparison, Linear Regression emerged as the most suitable approach for this task.<br>

#### Limitations
While the models performed well within the scope of this project, several limitations should be acknowledged:
- **Limited feature set**
The models rely on a small number of macroeconomic variables. Mortgage rates may also be influenced by factors not included here, such as lender competition, regulatory changes, or market sentiment.
- **Short-term forecasting focus**
The approach used for predicting future values relies on recursive forecasting, which can accumulate error over longer horizons. As a result, the model is best suited for short-term forecasts rather than long-term projections.
- **Assumption of linearity**
Linear Regression assumes stable linear relationships over time. Structural changes in the economy or sudden market shocks may reduce model accuracy if these relationships change.
- **No external forward-looking data**
Future values of explanatory variables were not forecast independently. Instead, recently observed or predicted values were reused, which limits realism in longer-range predictions.
- **Single-country and product scope**
The analysis focuses exclusively on UK mortgage rates at 75% LTV and may not generalise to other markets or mortgage products.

#### Future Work
Potential extensions of this project include:
- Incorporating additional macroeconomic and financial indicators
- Forecasting explanatory variables separately before predicting mortgage rates
- Evaluating performance during periods of economic stress
- Extending the framework to other loan-to-value ratios or international markets
